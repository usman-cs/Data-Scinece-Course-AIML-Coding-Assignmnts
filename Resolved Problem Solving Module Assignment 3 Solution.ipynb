```python
def is_prime(num,divisor=2):
    flag=False
    if num==1:
        return False
    if divisor==num:
        return True
    rem=num%divisor
    if rem==0:
        return False
    r=is_prime(num,divisor+1)
    if rem!=0:
        flag=r
        return flag
```

```python
def even_prime_list_creation(l):
    even=[]
    prime=[]
    if len(l)==0:
        return even,prime
    else:
        number=l[0]
        if number%2==0:
            even.append(number)
            if number==2:
                prime.append(number)
        elif (is_prime(number) and number!=1):
            prime.append(number)
        r=even_prime_list_creation(l[1:])
        even.extend(r[0])
        prime.extend(r[1])
        return even,prime
```

```python
import numpy as np
random_integers = np.random.randint(low=0, high=1500, size=(1000,)) 
evenNumberList,PrimeNumberList=even_prime_list_creation(random_integers)
print('Even Number List: ',evenNumberList)
print('\n\nPrime Number List: ',PrimeNumberList)
```
